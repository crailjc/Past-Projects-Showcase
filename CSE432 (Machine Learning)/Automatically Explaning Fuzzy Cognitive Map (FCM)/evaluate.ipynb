```python
from evaluate import *
import json
from tabulate import tabulate

f = open('evaluate/babbage_orig.json')
json_data = json.load(f)
scores = dict(
    b1=list(),
    b2=list(),
    b3=list(),
    b4=list(),
    m=list(),
)

for text in json_data:
    expected, obtained = text['expected'], text['generated']
    scores['b1'].append(evaluate_bleu1([expected], obtained))
    scores['b2'].append(evaluate_bleu2([expected], obtained))
    scores['b3'].append(evaluate_bleu3([expected], obtained))
    scores['b4'].append(evaluate_bleu4([expected], obtained))
    scores['m'].append(evaluate_meteor([expected], obtained))

# evaluate_bleurt(references, candidates)
print("=========Scores=========")
data = [
    ['BLEU-1', sum(scores['b1'])/len(scores['b1'])],
    ['BLEU-2', sum(scores['b2'])/len(scores['b2'])],
    ['BLEU-3', sum(scores['b3'])/len(scores['b3'])],
    ['BLEU-4', sum(scores['b4'])/len(scores['b4'])],
    ['METEOR', sum(scores['m'])/len(scores['m'])],
]
print(tabulate(data, headers=["Score", "Value"], tablefmt="simple"))
```
